```python
import pandas as pd

# Load Titanic — pick ONE option:
# A) Google Drive:
# from google.colab import drive; drive.mount('/content/drive')
# df = pd.read_csv('/content/drive/MyDrive/ML-30/01/train.csv')
# B) kagglehub:
# import kagglehub, os; p = kagglehub.competition_download('titanic'); df = pd.read_csv(os.path.join(p,'train.csv'))
# C) no accounts:
import seaborn as sns
df = sns.load_dataset('titanic')
df = df.rename(columns={'survived':'Survived','pclass':'Pclass','sex':'Sex','age':'Age','fare':'Fare'})
df.head()
```

```python
# EXAMPLE — prepare features + train
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
df['Sex'] = df['Sex'].map({'male':0,'female':1})
df['Age'] = df['Age'].fillna(df['Age'].median())
X = df[['Sex','Pclass','Age','Fare']]; y = df['Survived']
Xtr, Xte, ytr, yte = train_test_split(X, y, random_state=42)
model = LogisticRegression(max_iter=1000).fit(Xtr, ytr)
proba = model.predict_proba(Xte)[:,1]   # probability of survival
# print(proba)
```

```python
# TODO — print: confusion_matrix, classification_report, roc_auc_score
# imports: from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
# print(yte)
pred = model.predict(Xte)
cm = confusion_matrix(yte, pred)
print(cm)

cr = classification_report(yte, pred)
print(cr)

ras = roc_auc_score(yte, proba)
print(ras)
```

```python
# TODO — threshold experiment: classify with threshold 0.3 instead of 0.5
from sklearn.metrics import precision_score, recall_score

# Predictions for threshold 0.5
# print(proba)
pred_05 = (proba >= 0.5).astype(int)
# print(pred_05)
precision_05 = precision_score(yte, pred_05)
recall_05 = recall_score(yte, pred_05)

# Predictions for threshold 0.3
pred_03 = (proba >= 0.3).astype(int)
precision_03 = precision_score(yte, pred_03)
recall_03 = recall_score(yte, pred_03)

print(f"Threshold 0.5: Precision = {precision_05:.3f}, Recall = {recall_05:.3f}")
print(f"Threshold 0.3: Precision = {precision_03:.3f}, Recall = {recall_03:.3f}")
```

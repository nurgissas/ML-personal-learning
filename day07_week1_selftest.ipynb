### Q1 — Explain bias vs variance, and how you detect each from train/val scores.

bias occurs when model is too simple (underfit), variance occurs when model is too complicated (overfit on training data, thus can't predict accurately on validation data)

fixes for bias: more complex model (more features)
fixes for variance: more data, regularization (L1, L2), make the model simpler

### Q2 — Precision vs recall: define both and give one scenario where each is the priority.

precision - % of correctly predicted positives out of all positively predicted values (TP/(TP+FP)): when having false positive is costly, damaging: email spam filtering (not spam but identified as spam is better rather than letting few spams into the inbox(not spam))

recall - % of correctly predicted positives out of all actually positive values
(TP/(TP+FN)): when having false negative is costly, highly expensive: cancer detection (sick patients -> predicted as healthy is very costly)

### Q3 — Bagging vs boosting in two sentences.
Bagging (random forest) - train many models in parallel on bootstrapped samples, average them -> it reduces the variance

Boosting (XGBoost) - train sequentially, each model corrects the errors made by previous model -> reduces bias.
Boosting can easily overfit on the training data

### Q4 — What is data leakage, and what specifically does a Pipeline do to prevent it?

Data leakage is validation or testing data leaks into training data and validation data is partially imputed, altered or used to train the model. E.g. validation data is being used to calculate mean for imputing etc.

Using pipeline means applying the pipeline only on the training data. Pipeline can impute the missing features (num and categorical features) and prevents validation data from affecting the model

it goes like this:
Full dataset -> splits (training data) (validation data) -> pipeline.fit_transform(training data), pipeline.transform(validation data)

### Q5 — Why is accuracy misleading on imbalanced data, and what do you use instead?

accuracy is (correct predictions)/(total predictions)

aka TP+TN / (all predictions)

% of correctly predicted values

use accuracy when the data is balanced on both negative and positive
(e.g. data actual split 50/50 on cats, dogs image classification)
problem when you have to classify very rare case (medical disease), if model predicts breast cancer with accuracy of 99%, it can be misleading (since there are less cases of people having cancer unless the datasplit is 50 of cancer patients and 50 of healthy)

```python
# Q6 — FROM MEMORY: load any sklearn dataset, split, fit a model, print a test metric.
# If you can type this without hesitation, Week 1 did its job.
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklear.metrics import mean_absolute_error

tr = pd.read_csv() #import the dataset csv
X = tr.drop() # drop the actual y
y = tr['column to predict']

# separate the X's columns into num and col types
num_cols = X.select_dtypes('number').columns
cat_cols = X.select_dtypes('object').columns

pre = ColumnTransformer([
    ('num', SimpleImputer(strategy="median"), num_cols),
    ('cat', Pipeline([
        ('imp', SimpleImputer(strategy='most_frequent')),
        ('oh',  OneHotEncoder(handle_unknown='ignore')),
    ]), cat_cols),
])

# split into training data/testing data, can use cross val in here
Xtr, Xte, ytr, yte = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state=0)
param_1 = [('pre', pre), ('xgb', XGBRegressor(n_estimators=500, max_depth=3))]
pipe_1 = Pipeline(param_1)
pipe_1.fit(Xtr, ytr)
pred_1 = pipe_1.predict(Xte)
print(mean_absolute_error(yte, pred_1))
```
